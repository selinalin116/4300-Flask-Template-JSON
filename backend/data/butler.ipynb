Load in libraries

```python
import pandas
```

Load in JSONs

```python
ar = pandas.read_json("recipes_raw_nosource_ar.json")
epi = pandas.read_json("recipes_raw_nosource_epi.json")
fn = pandas.read_json("recipes_raw_nosource_fn.json")
```

```python
lengths = []

for df in [ar.loc['ingredients'], epi.loc['ingredients'], fn.loc['ingredients']]:
    for row in df:
        if type(row) is list:
            lengths.append(len(row))
```

```python
print(len(lengths))
```

```python
df = pandas.DataFrame({"Recipe Ingredient Counts":lengths})
hist = df.hist(bins=20)
```
