Load in libraries

```python
import pandas
import matplotlib.pyplot as plt
```

Load in JSONs

```python
ar = pandas.read_json("recipes_raw_nosource_ar.json")
epi = pandas.read_json("recipes_raw_nosource_epi.json")
fn = pandas.read_json("recipes_raw_nosource_fn.json")
```

```python
lengths = []
title_words = []

for df in [ar.loc['ingredients'], epi.loc['ingredients'], fn.loc['ingredients']]:
    for row in df:
        if type(row) is list:
            lengths.append(len(row))

for df in [ar.loc['title'], epi.loc['title'], fn.loc['title']]:
    for row in df:
        if type(row) is str:
            title_words.append(len(row.split(" ")))
```

```python
print(len(lengths))
print(title_words)
```

```python
df = pandas.DataFrame({"Recipe Ingredient Counts":lengths})
hist = plt.hist(df, bins=20, edgecolor='black', linewidth=1.2)
plt.title("Counts of Recipes by Ingredient Count")
plt.xlabel("Number of Ingredients")
plt.ylabel("Count")
```

```python
df = pandas.DataFrame({"Number of Words in the Recipe Title":title_words})
hist = plt.hist(df, range=[0,10], edgecolor='black', linewidth=1.2)
plt.title("Counts of Recipes by Title Word Count")
plt.xlabel("Number of Words")
plt.ylabel("Count")
```
